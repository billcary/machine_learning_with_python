```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
```

```python
import pandas as pd
from pandas import DataFrame
from pandas import Series
```

```python
# we load the data with load_iris from sklearn
data = load_iris()
features = data['data']
feature_names = data['feature_names']
target = data['target']
```

```python
irisfeatures = DataFrame(data=features, columns=feature_names)
```

```python
irisfeatures[:5]
```

```python
iristargets = DataFrame(data=target)
```

```python
iristargets[:5]
```

```python
iristargets[0].describe()
```

```python
# Merge the feature and target dataframes (add target as a column on the end of features
irisdata = pd.merge(left=irisfeatures, right=iristargets, left_index=True, right_index=True)
```

```python
irisdata[:5]
```

```python
# Rename last column
irisdata.rename(columns={0:'class'}, inplace=True)
```

```python
irisdata[:5]
```

```python
irisdata.head()
```

```python
from pandas.tools.plotting import scatter_matrix
```

```python
colors = irisdata['class'].replace(0, 'r').replace(1, 'g').replace(2, 'b')
```

```python
scatter_matrix(irisdata, diagonal='kde', alpha=0.2, color=colors, figsize = (25, 20))
```
